```python
sum = 0
print("Question6 response:")
for i in l:
    if type(i) == list or type(i) == tuple or type(i) == set:
        for k in i:
            if type(k)==int:
                print(k)
                a = k
                sum = a+sum
    if type(i) == dict:
        for u,v in i.items():
            if type(u)==int or type(v)==int:
                print(u)
                print(v)
                b = u
                c = v
                sum = b+c+sum
print()
print("Question7 response:\n","sum of all numbers is ",sum,sep="")
```

```python
print("Question10 response: Number of occurances of all the data")
l_unwrapped = []
for i in l:
    if type(i) == list or type(i) == tuple or type(i) == set:
        for k in i:
            l_unwrapped.append(k)
    if type(i) == dict:
        for u,v in i.items():
            l_unwrapped.append(u)
            l_unwrapped.append(v)
#print(l_unwrapped)
for i in set(l_unwrapped):
    print (i,":",l_unwrapped.count(i))
```
