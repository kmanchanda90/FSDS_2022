```python
def outlier_threshold(df,variable):
    q1 = df[variable].quantile(0.25)
    q2 = df[variable].quantile(0.75)
    iqr = q2 - q1
    up_limit = q2 + (1.5*iqr)
    low_limit = q1 - (1.5*iqr)
    return up_limit, low_limit
```
